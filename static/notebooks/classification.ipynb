### Classification Analysis (SVM)

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
var_entity_1 = "bfd"
```

```python
var_entity_2 = "entity 1"
```

```python
%matplotlib inline
import numpy as np
import matplotlib
from sklearn import svm

np.random.seed(0)
X = np.r_[np.random.randn(1500, 2) - [2, 2], np.random.randn(1500, 2) + [2, 2]]
Y = [0] * 1500 + [1] * 1500

plt.figure(figsize=(15,10))
plt.xlabel(var_entity_1)
plt.ylabel(var_entity_2)
clf2 = svm.LinearSVC(C=1).fit(X, Y)

# get the separating hyperplane
w = clf2.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - (clf2.intercept_[0]) / w[1]

# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx2, yy2 = np.meshgrid(np.arange(x_min, x_max, .2),
                     np.arange(y_min, y_max, .2))
Z = clf2.predict(np.c_[xx2.ravel(), yy2.ravel()])

Z = Z.reshape(xx2.shape)

cmap_1 = matplotlib.colors.LinearSegmentedColormap.from_list("",["#ffffff", "#ffffff", "#ffffff", "#ffffff", "#ffffff"])
cmap_2 = matplotlib.colors.LinearSegmentedColormap.from_list("",["#009FDB", "#00719C", "#00678F", "#005575", "#00394F"])

plt.contourf(xx2, yy2, Z, cmap=cmap_1, alpha=0.3)
color_dict = {0 : '#00535a',1 : '#b4126b', 2: "#D4BF2A"}
cm = [color_dict[x] for x in Y]
plt.scatter(X[:, 0], X[:, 1], c=cm , s=60, alpha = 0.6)
plt.plot(xx,yy,"--k", lw=4, c= "black")

ax = plt.axis([x_min, x_max,y_min, y_max])
#plt.show()
```
