### Clustering Analysis

```python
import numpy as np
import pandas as pd 
import sklearn
import matplotlib.pyplot as plt
```

```python
var_entity_1 = "groups"
```

```python
var_entity_2 = "entity 1"
```

```python
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.patches as mpatches


plt.figure(figsize = (15,10))
plt.xlabel('x1')
plt.ylabel('x2')

n_samples = 1500
random_state = 20
X, y = make_blobs(n_samples=3500, n_features = 2, cluster_std=1.8,  centers = 3, random_state=random_state)

kmeans = KMeans(n_clusters=3, n_jobs = -1).fit(X)

recs = []
classes = [var_entity_1, var_entity_2, "entity n"]
colors = ["#00535a", "#b4126b", "#D4BF2A"]
for i in range(0,len(colors)):
    recs.append(mpatches.Rectangle((0,0),2,2,fc=colors[i]))
plt.legend(recs,classes,loc=1)

color_dict = {0 : '#00535a',1 : '#b4126b', 2: "#D4BF2A"}
cm = [color_dict[x] for x in kmeans.labels_]
ax = plt.scatter(X[:, 0], X[:, 1], c=cm, alpha = 0.6, s = 80)
```
